```python
board = [" " for _ in range(9)]

def show_board():
    for i in range(0, 9, 3):
        print(board[i] + " | " + board[i+1] + " | " + board[i+2])
        if i < 6:
            print("--+---+--")
def check_win(player):
    wins = [(0,1,2), (3,4,5), (6,7,8),
            (0,3,6), (1,4,7), (2,5,8),
            (0,4,8), (2,4,6)]
    for a, b, c in wins:
        if board[a] == board[b] == board[c] == player:
            return True
    return False
def is_tie():
    return " " not in board
def best_move():
    best_score = -1000
    move = 0
    for i in range(9):
        if board[i] == " ":
            board[i] = "O"
            score = minimax(False)
            board[i] = " "
            if score > best_score:
                best_score = score
                move = i
    board[move] = "O"
def minimax(is_computer):
    if check_win("O"):
        return 1
    if check_win("X"):
        return -1
    if is_tie():
        return 0

    if is_computer:
        best = -1000
        for i in range(9):
            if board[i] == " ":
                board[i] = "O"
                score = minimax(False)
                board[i] = " "
                best = max(score, best)
        return best
    else:
        best = 1000
        for i in range(9):
            if board[i] == " ":
                board[i] = "X"
                score = minimax(True)
                board[i] = " "
                best = min(score, best)
        return best

# Start the game
print("Welcome to Tic-Tac-Toe!")
print("You are X and computer is O")
show_board()

while True:
    try:
        user = int(input("Pick a place (1-9): ")) - 1
        if board[user] != " ":
            print("That place is already taken. Try again.")
            continue
    except:
        print("Please enter a number between 1 and 9.")
        continue

    board[user] = "X"

    if check_win("X"):
        show_board()
        print("You won!")
        break

    if is_tie():
        show_board()
        print("It's a draw.")
        break

    best_move()

    show_board()

    if check_win("O"):
        print("Computer won.")
        break

    if is_tie():
        print("It's a draw.")
        break
```
